## Dataset Overview
#### Rows (Samples): 10,000
#### Columns (Attributes): 10
#### Purpose: This dataset appears to relate to student academic profiles and their placement outcomes.

##### Attribute (Column) Information
### Column Name		            Description
1. College_ID		           : Unique identifier for each student's college.
2. IQ		                   : Student's IQ score.
3. Prev_Sem_Result		       : GPA or percentage from the previous semester.
4. CGPA		                   : Cumulative Grade Point Average.
5. Academic_Performance		   : Presumably a rating (1â€“10) of academic performance.
6. Internship_Experience	   : Whether the student has internship experience.
7. Extra_Curricular_Score	   : Score reflecting participation in extracurricular activities.
8. Communication_Skills		   : Score for communication abilities.
9. Projects_Completed		   : Number of academic or personal projects completed.
10. Placement		           : Whether the student was placed in a job.
